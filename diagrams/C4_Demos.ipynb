## Contexto

```python
from diagrams import Diagram
from diagrams.c4 import Person, System, Relationship

with Diagram("Biblioteca - C4 Contexto", show=False, direction="TB") as diagram_demo:
    usuario = Person(name="Usuario", description="Busca y solicita libros")
    bibliotecario = Person(name="Bibliotecario", description="Administra libros y préstamos")
    sistema = System(name="Sistema de Biblioteca", description="Gestiona libros, préstamos y usuarios")
    email = System(name="Servicio de Email", description="Envía recordatorios de préstamos")

    usuario >> Relationship("Consulta y solicita libros") >> sistema
    bibliotecario >> Relationship("Gestiona catálogo y préstamos") >> sistema
    sistema >> Relationship("Envía recordatorios") >> email

diagram_demo
```

## Contenedor

```python
from diagrams import Diagram
from diagrams.c4 import Person, Container, Database, SystemBoundary, Relationship

with Diagram("Sistema de Biblioteca - C4 Container", show=False, direction="TB") as diagram_demo:
    usuario = Person(name="Usuario", description="Accede desde el navegador")

    with SystemBoundary("Sistema de Biblioteca"):
        web = Container(name="Web App", technology="React", description="Interfaz web para usuarios")
        api = Container(name="API REST", technology="Node.js/.NET", description="Lógica del sistema de biblioteca")
        db = Database(name="Base de Datos", technology="Azure SQL", description="Almacena libros, préstamos, usuarios")
        email = Container(name="Servicio de Email", technology="Azure Communication", description="Envío de correos")

    usuario >> Relationship("Usa desde navegador") >> web
    web >> Relationship("Solicita datos") >> api
    api >> Relationship("Lectura/Escritura") >> db
    api >> Relationship("Envía correos") >> email

diagram_demo
```
