```python
from diagrams import Diagram, Cluster
from diagrams.azure.compute import AppServices
from diagrams.azure.network import LoadBalancers
from diagrams.azure.web import AppServices, APIConnections
from diagrams.azure.database import SQLDatabases
from diagrams.azure.identity import Users


with Diagram("Biblioteca en la Nube", show=False, direction="LR") as diagram_demo:
    
    users = Users("Biblioteca Usuarios")
    web_app = AppServices("Biblioteca Web App")
    lb = LoadBalancers("Load Balancer")

    users >> web_app >> lb

    with Cluster("Capa de servicios"):
        api1 = AppServices("API Servicio 1")
        api2 = AppServices("API Servicio 2")
        api3 = AppServices("API Servicio N")

        lb >> [api1, api2, api3]

    db_cluster = SQLDatabases("SQL Cluster\n\n(Libros, Usuarios, Prstamos)")

    [api1, api2, api3] >> db_cluster

diagram_demo
```
